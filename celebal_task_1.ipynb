NumPy

```python
import numpy as np

# Creating an array
arr = np.array([1, 2, 3, 4, 5])
print("Original Array:", arr)
```

```python
# Accessing array elements
print("First element:", arr[0])
print("Last element:", arr[-1])
print("Slice of array:", arr[2:4])
```

```python

# Shape and dimensions of an array
print("Shape of array:", arr.shape)
print("Number of dimensions:", arr.ndim)
print("Size of array:", arr.size)
```

```python

# Reshaping an array
reshaped_arr = arr.reshape(5, 1)
print("Reshaped Array:\n", reshaped_arr)
```

```python

# Arithmetic operations on arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
addition = np.add(a, b)
subtraction = np.subtract(a, b)
multiplication = np.multiply(a, b)
division = np.divide(a, b)
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

```python

# Statistical functions
data = np.array([10, 20, 30, 40, 50])
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python

# Random number generation
random_nums = np.random.rand(5)
print("Random numbers:", random_nums)
```

```python

# Array manipulation
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
concatenated_arr = np.concatenate((arr1, arr2), axis=0)
transposed_arr = np.transpose(arr1)
print("Concatenated Array:\n", concatenated_arr)
print("Transposed Array:\n", transposed_arr)
```

Pandas

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Emma', 'Andrew', 'Olivia'],
        'Age': [25, 28, 22, 30],
        'City': ['New York', 'London', 'Paris', 'Sydney']}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python

# Accessing DataFrame columns
print("Column 'Name':\n", df['Name'])
print("Column 'Age':\n", df.Age)
```

```python

# Accessing DataFrame rows
print("Row at index 1:\n", df.loc[1])
print("Rows from index 2 to 3:\n", df.loc[2:3])
```

```python

# Descriptive statistics
print("Statistics:\n", df.describe())
```

```python

# Sorting DataFrame
sorted_df = df.sort_values(by='Age')
print("Sorted DataFrame:\n", sorted_df)
```

```python

# Filtering rows
filtered_df = df[df['Age'] > 25]
print("Filtered DataFrame:\n", filtered_df)
```

```python

# Adding a new column
df['Gender'] = ['Male', 'Female', 'Male', 'Female']
print("DataFrame with new column:\n", df)
```

```python

# Grouping and aggregating data
grouped_df = df.groupby('Gender').mean()
print("Grouped DataFrame:\n", grouped_df)
```

```python

# Handling missing values
df_with_missing = pd.DataFrame({'A': [1, 2, np.nan, 4],
                                'B': [5, np.nan, 7, 8],
                                'C': [9, 10, 11, np.nan]})
print("DataFrame with missing values:\n", df_with_missing)
filled_df = df_with_missing.fillna(0)
print("DataFrame with filled missing values:\n", filled_df)
```

```python

# Reading and writing data
df.to_csv('data.csv', index=False)
read_df = pd.read_csv('data.csv')
print("Read DataFrame:\n", read_df)
```

MatplotLib

```python
import matplotlib.pyplot as plt
import numpy as np

# Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python

# Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python

# Bar plot
x = ['A', 'B', 'C', 'D']
y = [10, 20, 15, 25]
plt.bar(x, y)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python

# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 15]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python

# Box plot
data = np.random.randn(1000)
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data')
plt.show()
```

```python

# Heatmap
data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```
